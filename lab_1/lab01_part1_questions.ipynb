## Exercise 1

$$  
y = x^Tx,  \quad x \in \mathbb{R}^N 
$$

$$
\frac{dy}{dx} = 2x^T
$$ 

$ y =  x^Tx = \sum_{i=1}^n x_i^2 $ 

for each  $x_k:  \frac{\partial y}{\partial x_k}  = \frac{\partial}{\partial x_k} (\sum_{i=1}^n x_i^2) = \frac{\partial}{\partial x_k} (x_k^2) = 2x_k $ 

then vector looks like: $ \frac{\partial y}{\partial x} = [ \frac{\partial y}{\partial x_1} ... \frac{\partial y}{\partial x_n} ]  = [2x_1 ... 2x_n] = 2x^T $

## Exercise 2

$ y = tr(AB) \quad A,B \in \mathbb{R}^{N \times N} $

$$
\frac{dy}{dA} = B^T
$$

$(AB)_{ij} = \sum_{k=1}^n a_{ik} b_{kj}$

$y = tr(AB) = \sum_{i=1}^n \sum_{j=1}^n a_{ij} b_{ji}$

lets define partial differentiation for some 
$A_{kl}$: 
$(\frac{\partial y}{\partial A})_{kl} = \frac{\partial y}{\partial A_{kl}}$

let's see how our sum looks like:
$ y = A_{11}B_{11} + A_{12}B_{21} + ... + A_{kl}B_{lk} + ... $

we can see that: $ \frac{\partial (A_{kl}B_{lk})}{\partial A_{kl}} = B_{lk} = (B_{kl})^T$

when we differentiate by $A_{kl}$ all other monomials = 0 where i,j $\not=$ k,l 

therefore: $ \frac{dy}{dA} = B^T $
